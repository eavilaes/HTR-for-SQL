```
import os

os.mkdir('./dataset')
os.mkdir('./dataset/img')

line_n = 0
with open('PerfectSQL_2.sql', 'r') as f:
  with open('words.txt', 'w') as w:
    for line in f.readlines():
     
      # Create group folder
      group_n = line_n // 30 # 15 Groups of 30 lines (for 446 sentences)
      try:
        os.mkdir('./dataset/img/s' + "{0:0=2d}".format(group_n))
      except:
        pass
      
      # Create line folder
      NEW_DIR = './dataset/img/s' + "{0:0=2d}".format(group_n) + '/s' + "{0:0=2d}".format(group_n) + '-l' + "{0:0=3d}".format(line_n)
      try:
        os.mkdir(NEW_DIR)
      except:
        pass

      word_n = 0
      for word in line.split():
        for hws in range(5): # Do it for each handwriting style
          # Move files from line to that folder and rename them
          try:
            ORIGINAL_FILE = './content/output/line' + str(line_n) + '/word' + str(word_n) + '_img' + str(hws) + '.png'
            FILENAME = 's' + "{0:0=2d}".format(group_n) + '-l' + "{0:0=3d}".format(line_n) + '-' + "{0:0=2d}".format(hws) + '-' + "{0:0=2d}".format(word_n)
            NEW_FILE = NEW_DIR + '/' + FILENAME + '.png'
            os.rename(ORIGINAL_FILE, NEW_FILE)
          except:
            pass

          # Write the word into the words file
          w.write(FILENAME + " ok 000 000 000 000 00 AA " + word + '\n')

        word_n += 1 # for word in line
      line_n += 1 # for line in f
```
